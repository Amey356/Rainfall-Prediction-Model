```python
df.drop(['Evaporation','Sunshine','Cloud9am','Cloud3pm','Date'], axis = 1 , inplace = True)         
```

```python
# For numeric columns ---  use median to fill null values
df['MinTemp'].fillna(df['MinTemp'].median(), inplace=True)
df['MaxTemp'].fillna(df['MaxTemp'].median(), inplace=True)
df['Rainfall'].fillna(df['Rainfall'].median(), inplace=True)
df['WindGustSpeed'].fillna(df['WindGustSpeed'].median(), inplace=True)
df['WindSpeed9am'].fillna(df['WindSpeed9am'].median(), inplace=True)
df['WindSpeed3pm'].fillna(df['WindSpeed3pm'].median(), inplace=True)
df['Humidity9am'].fillna(df['Humidity9am'].median(), inplace=True)
df['Humidity3pm'].fillna(df['Humidity3pm'].median(), inplace=True)
df['Pressure9am'].fillna(df['Pressure9am'].median(), inplace=True)
df['Pressure3pm'].fillna(df['Pressure3pm'].median(), inplace=True)
df['Temp9am'].fillna(df['Temp9am'].median(), inplace=True)
df['Temp3pm'].fillna(df['Temp3pm'].median(), inplace=True)
df['RISK_MM'].fillna(df['RISK_MM'].median(), inplace=True)

# For categorical columns, use mode to fill null values
df['Location'].fillna(df['Location'].mode()[0], inplace=True)
df['WindGustDir'].fillna(df['WindGustDir'].mode()[0], inplace=True)
df['WindDir9am'].fillna(df['WindDir9am'].mode()[0], inplace=True)
df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0], inplace=True)
df['RainToday'] = df['RainToday'].fillna(df['RainToday'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)
```

```python
# Data Balanced


import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,5))
df.RainTomorrow.value_counts(normalize = True).plot(kind='bar', color= ['skyblue','navy'], alpha = 0.9, rot=0)
plt.title('RainTomorrow Indicator No(0) and Yes(1) in the Imbalanced Dataset')
plt.show()
```

```python
df['RainTomorrow'].value_counts()
```

```python
# Under sampling

no_cat = df[df['RainTomorrow'] == 0].sample(31877)
no_cat
```

```python
yes_cat = df[df['RainTomorrow'] == 1]
yes_cat
```

```python
df_resampled = pd.concat([no_cat,yes_cat], axis = 0)
df_resampled
```

```python
cat = []
num = []


for i in df_resampled.columns:
    if df_resampled[i].dtype == 'O':
        cat.append(i)
    else:
        num.append(i)
```

```python
cat
```

```python
num
```

```python
for i in df_resampled.columns:
    if df_resampled[i].dtype == 'O':
        print(df[i].value_counts())
        print('--------------------------')
        
```

```python
for i in cat:
    vc = df_resampled[i].value_counts()
    plt.bar(vc.index, height = vc , color = 'b')
    plt.xlabel(i)
    plt.xticks(rotation= 90)
    plt.ylabel('Count')
    plt.show()
```

```python
# Numerical

for i in num:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(df_resampled[i],color = 'c')
    ax[1].boxplot(df_resampled[i])
    plt.suptitle(i)
    plt.show()
```

```python
# iqr winsorization --
# replace lower outliers with lower fence
# replace upper outliers with upper fence


for i in num:
    q1 = df_resampled[i].quantile(0.25)
    q3 = df_resampled[i].quantile(0.75)
    iqr = q3 - q1
    lf = q1 - 1.5*iqr
    uf = q3 + 1.5*iqr
    df_resampled.loc[df_resampled[i]<lf, i] = lf
    df_resampled.loc[df_resampled[i]>uf, i] = uf
```

## Encoding

```python

from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

df_resampled['Location'] = label_encoder.fit_transform(df_resampled['Location'])
df_resampled['WindGustDir'] = label_encoder.fit_transform(df_resampled['WindGustDir'])
df_resampled['WindDir9am'] = label_encoder.fit_transform(df_resampled['WindDir9am'])
df_resampled['WindDir3pm'] = label_encoder.fit_transform(df_resampled['WindDir3pm'])

print(df_resampled.head())
```

```python
df_resampled
```

```python
X_df = df_resampled.drop(['RainTomorrow'], axis = 1)
y = df_resampled['RainTomorrow']
X_df
```

```python
y
```

```python
X = X_df.values
```

```python
Y = y.values
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, confusion_matrix

# Training the model
model = LogisticRegression()
model.fit(x_train, y_train)

# Predicting on training and testing data
y_train_predicted = model.predict(x_train)
y_test_predicted = model.predict(x_test)

# Training Performance
print("Training Performance: ")
print("Training accuracy: ", accuracy_score(y_train, y_train_predicted))
print("Training precision: ", precision_score(y_train, y_train_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Training recall: ", recall_score(y_train, y_train_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_train, y_train_predicted))
print("Training classification report: ", classification_report(y_train, y_train_predicted))

# Testing Performance
print("Testing Performance: ")
print("Testing accuracy: ", accuracy_score(y_test, y_test_predicted))
print("Testing precision: ", precision_score(y_test, y_test_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Testing recall: ", recall_score(y_test, y_test_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_test, y_test_predicted))
print("Testing classification report: ", classification_report(y_test, y_test_predicted))
```

# Decision tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Initialize and train the decision tree classifier
dtree = DecisionTreeClassifier(criterion='gini', max_depth= 3 , min_samples_leaf=1, min_samples_split=2)
dtree.fit(x_train, y_train)

# Predict on training and testing data
y_train_predicted = dtree.predict(x_train)
y_test_predicted = dtree.predict(x_test)

# Training Performance
print("Training Model Performance:")
print("Training Accuracy:", accuracy_score(y_train, y_train_predicted))
print("Training Precision:", precision_score(y_train, y_train_predicted, average='binary'))  # Use 'macro' or 'weighted' for multi-class
print("Training Recall:", recall_score(y_train, y_train_predicted, average='binary'))  # Use 'macro' or 'weighted' for multi-class
print("Training Confusion Matrix:\n", confusion_matrix(y_train, y_train_predicted))
print("Training Classification Report:\n", classification_report(y_train, y_train_predicted))

# Testing Performance
print("\nTesting Model Performance:")
print("Testing Accuracy:", accuracy_score(y_test, y_test_predicted))
print("Testing Precision:", precision_score(y_test, y_test_predicted, average='binary'))  # Use 'macro' or 'weighted' for multi-class
print("Testing Recall:", recall_score(y_test, y_test_predicted, average='binary'))  # Use 'macro' or 'weighted' for multi-class
print("Testing Confusion Matrix:\n", confusion_matrix(y_test, y_test_predicted))
print("Testing Classification Report:\n", classification_report(y_test, y_test_predicted))
```

# GridSearch CV - Decision Tree Classifier


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the Decision Tree Classifier
dtree = DecisionTreeClassifier()

# Define the parameter grid to search over
param_grid = {
    'max_depth': [3, 5, 10,15,20, None],  # Depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],  # Minimum samples required at a leaf node
    'criterion': ['gini', 'entropy']  # Split criterion
}

# Initialize the GridSearchCV
grid_search = GridSearchCV(estimator=dtree, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)

# Fit the GridSearchCV to the training data
grid_search.fit(x_train, y_train)

# Print the best hyperparameters found
print(f"Best Hyperparameters: {grid_search.best_params_}")

# Get the best model from GridSearchCV
best_dtree_model = grid_search.best_estimator_

# Predict on training data using the best model
y_train_predicted = best_dtree_model.predict(x_train)

# Evaluate on training data
print("\nTraining Model Performance:")
print("Training Accuracy:", accuracy_score(y_train, y_train_predicted))
print("Training Classification Report:\n", classification_report(y_train, y_train_predicted))
print("Training Confusion Matrix:\n", confusion_matrix(y_train, y_train_predicted))

# Predict on testing data using the best model
y_test_predicted = best_dtree_model.predict(x_test)

# Evaluate on testing data
print("\nTesting Model Performance:")
print("Testing Accuracy:", accuracy_score(y_test, y_test_predicted))
print("Testing Classification Report:\n", classification_report(y_test, y_test_predicted))
print("Testing Confusion Matrix:\n", confusion_matrix(y_test, y_test_predicted))
```

# AdaBoostClassifier
